### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "LocalAuthorityName": "London",  # Filter by Local Authority
    "RatingValue": {"$gte": 4}        # Filter for RatingValue greater than or equal to 4
}

# Count the number of documents matching the query
london_rating_count = establishment.count_documents(query)
print(f"Number of establishments in London with RatingValue >= 4: {london_rating_count}")

# Find the first document that matches the query
first_document = establishment.find_one(query)

# Display the first document if it exists
if first_document:
    print("First establishment in London with RatingValue >= 4:")
    pprint(first_document)
else:
    print("No establishments found in London with RatingValue >= 4.")
```

```python
results = list(establishment.find(query))

# Check the structure of the results
if results and isinstance(results, list):
    # Convert to DataFrame
    df = pd.DataFrame(results)

    # Display the number of rows in the DataFrame
    print(f"Number of establishments in London with RatingValue >= 4: {len(df)}")

    # Display the first 10 rows of the DataFrame
    print("\nFirst 10 establishments in London with RatingValue >= 4:")
    print(df.head(10))
else:
    print("No valid data retrieved or data structure is not a list.")
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

atitude = 51.49014200
longitude = 0.08384000
degree_search = 0.01

# Define the query to find establishments
query = {
    "RatingValue": 5,  # Rating value must equal 5
    "geocode.latitude": {"$gte": atitude - degree_search, "$lte": atitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}

# Define the sort and limit
sort = [("scores.Hygiene", 1)]  # Sort by hygiene score (ascending)
limit = 5  # Limit to top 5 results

# Find the top 5 establishments
top_establishments = list(establishment.find(query).sort(sort).limit(limit))

# Print the results
if top_establishments:
    print(f"Top {limit} establishments with RatingValue of 5, sorted by lowest hygiene score:")
    for establishments in top_establishments:
        pprint(establishments)
else:
    print("No establishments found with the specified criteria.")
```

```python
# Convert result to Pandas DataFrame
if top_establishments:
    df = pd.DataFrame(top_establishments)
    
    # Display the number of rows in the DataFrame
    print(f"Number of establishments found: {len(df)}")
    
    # Display the first 10 rows of the DataFrame
    print("\nFirst 10 establishments:")
    print(df.head(10))
else:
    print("No establishments found with the specified criteria.")
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
#  Create the aggregation pipeline
pipeline = [
    {
        "$match": {
            "scores.Hygiene": 0  # Match establishments with hygiene score of 0
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",  # Group by Local Authority
            "count": {"$sum": 1}           # Count the number of establishments
        }
    },
    {
        "$sort": {
            "count": -1  # Sort by count in descending order
        }
    }
]

# Execute the aggregation pipeline
try:
    results = list(establishment.aggregate(pipeline))

    # Step 4: Print the number of documents in the result
    print(f"Number of Local Authorities with hygiene score of 0: {len(results)}")

    # Print the first 10 results
    print("\nFirst 10 results:")
    for result in results[:10]:
        pprint(result)

except Exception as e:
    print(f"An error occurred: {e}")
```
